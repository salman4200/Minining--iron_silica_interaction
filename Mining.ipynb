```python
from tqdm import tqdm
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import pandas as pd

# Load your dataset into 'mn' DataFrame
# Assuming 'mn' is your dataset DataFrame

# Interaction features
mn['iron_silica_interaction'] = mn['% Iron Feed'] * mn['% Silica Feed']

# Select features and target variable
X = mn[['% Iron Feed', '% Silica Feed', 'iron_silica_interaction']]
y = mn['% Iron Concentrate']  # Assuming '% Iron Concentrate' is your target variable

# Create a Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Perform cross-validation to evaluate the model
with tqdm(total=5, desc="Cross-Validation", position=0, leave=True) as pbar:
    scores = []
    for i, score in enumerate(cross_val_score(rf_model, X, y, cv=5, scoring='neg_mean_squared_error')):
        scores.append(-score)
        pbar.set_postfix({'fold': f'{i + 1}/5', 'MSE': f'{score:.2f}'})
        pbar.update(1)
    mean_mse = sum(scores) / len(scores)
    pbar.set_postfix({'Mean MSE': f'{mean_mse:.2f}'})
    pbar.update(1)

print(f"Mean Squared Error (Cross-Validated): {mean_mse}")

# Train the model on the full dataset
rf_model.fit(X, y)

# Get feature importances
importances = rf_model.feature_importances_
print("Feature Importances:", importances)

# Assuming 'test_df' is your test set DataFrame with the same columns as X
X_test_selected = mn[['% Iron Feed', '% Silica Feed', 'iron_silica_interaction']]

# Make predictions
y_pred = rf_model.predict(X_test_selected)
```

# Model Comparison

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Assuming you have trained and tested other models (Linear Regression and Gradient Boosting)
lr_model = LinearRegression()
lr_model.fit(X, y)
lr_y_pred = lr_model.predict(X)

gb_model = GradientBoostingRegressor()
gb_model.fit(X, y)
gb_y_pred = gb_model.predict(X)

# Calculate MAE and R-squared for each model
rf_mae = mean_absolute_error(y_true, y_pred_cv)
lr_mae = mean_absolute_error(y, lr_y_pred)
gb_mae = mean_absolute_error(y, gb_y_pred)

rf_r2 = r2_score(y_true, y_pred_cv)
lr_r2 = r2_score(y, lr_y_pred)
gb_r2 = r2_score(y, gb_y_pred)

# Visualization
models = ['Random Forest', 'Linear Regression', 'Gradient Boosting']
mae_scores = [rf_mae, lr_mae, gb_mae]  # MAE scores for each model
r2_scores = [rf_r2, lr_r2, gb_r2]  # R-squared scores for each model

plt.figure(figsize=(12, 6))

# MAE comparison
plt.subplot(1, 2, 1)
plt.bar(models, mae_scores, color='b')
plt.xlabel('Model')
plt.ylabel('MAE')
plt.title('MAE Comparison')

# R-squared comparison
plt.subplot(1, 2, 2)
plt.bar(models, r2_scores, color='g')
plt.xlabel('Model')
plt.ylabel('R-squared')
plt.title('R-squared Comparison')

plt.tight_layout()
plt.show()
```
